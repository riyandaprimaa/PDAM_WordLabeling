```python
import pandas as pd

dfs = [df1, df2, df3, df4, df5, df6]

# Gabungkan semua DataFrames menjadi satu
combined_df = pd.concat(dfs, ignore_index=True)

# Tampilkan hasilnya
combined_df = combined_df.dropna()
combined_df.isna().sum()
```

Modelling SVM (Support Vector Machine)

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline


# Pisahkan data menjadi data latih dan data uji
texts = combined_df['Comment']
labels = combined_df['Tag']

# Gunakan model pipeline dengan TF-IDF dan SVM
model = Pipeline([
    ('vectorizer', TfidfVectorizer()),
    ('classifier', SVC(kernel='linear'))
])

# Latih model
model.fit(texts, labels)

# Fungsi untuk menambahkan label pada setiap kata dalam kalimat
def label_tokens(sentence, model):
    tokens = sentence.split()
    labeled_tokens = []
    for token in tokens:
        prediction = model.predict([token])[0]
        labeled_tokens.append(f"{token} ({prediction})")
    return ' '.join(labeled_tokens)
```

Evaluation Model

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score

# Split the data into training and testing sets
texts_train, texts_test, labels_train, labels_test = train_test_split(texts, labels, test_size=0.2, random_state=42)

# Train the model on the training set
model.fit(texts_train, labels_train)

# Predict labels for the test set
predictions = model.predict(texts_test)

# Evaluate the model
accuracy = accuracy_score(labels_test, predictions)
classification_rep = classification_report(labels_test, predictions)

print(f"Accuracy: {accuracy}")
print("Classification Report:\n", classification_rep)
```
